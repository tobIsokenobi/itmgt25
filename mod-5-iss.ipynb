```python
import pandas as pd
import matplotlib as pldt
import requests
import time
import numpy as np

latlist = []
longlist = []
timestamplist = []
for i in range(100):
    a = requests.get("http://api.open-notify.org/iss-now.json")
    a = a.json()
    latitudept = a['iss_position']['latitude']
    longitudept = a['iss_position']['longitude']
    timestamp =  a['timestamp']
    latlist.append(latitudept)
    longlist.append(longitudept)
    timestamplist.append(timestamp)
    time.sleep(1)
boom = {'Timestamp': timestamplist, 'Longitude': longlist, 'Latitude': latlist}
df = pd.DataFrame(data=boom)
df
```

```python
fig, ax = plt.subplots( figsize=(13,13))
s = timestamplist[0]
e = timestamplist[99]
ax.scatter(df["Latitude"], df["Longitude"])
ax.xaxis.set_major_locator(plt.MaxNLocator(40))
ax.yaxis.set_major_locator(plt.MaxNLocator(40))
ax.set_xlabel("Latitude")
ax.set_ylabel("Longitude")
plt.xticks(rotation=45)
ax.set_title(f"Path of ISS for every 10 seconds from {s} to {e}")
```
